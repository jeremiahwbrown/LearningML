```python
class NeuralNetwork:

  def __init__(self, number_of_neurons, number_of_weights):
    self.size = number_of_neurons
    self.num_neurons = number_of_neurons
    self.hidden_layer = np.random.uniform(-.3, .3, (number_of_neurons, number_of_weights))
    self.gradients = np.zeros((number_of_neurons, number_of_weights))
    self.mults = np.zeros((number_of_weights, 1))
    self.output_layer = np.zeros((number_of_neurons, ))

  def feed_forward(self, input):
    self.mult = input
    sum = np.dot(self.hidden_layer, input)
    exp_sum  = np.exp(sum)
    softmax_sum = exp_sum/np.sum(exp_sum)
    self.output_layer = softmax_sum
    return softmax_sum

  def gradient_descent(self, loss):
    self.gradients += np.dot(loss.reshape((self.size, 1)), self.mult.reshape(1, self.num_neurons))

  def back_propigate(self, a):
    self.hidden_layer -= self.gradients * a
    self.gradients = np.zeros((self.size, self.num_neurons))
```

```python
class ProbabilisticGraph:

  def __init__(self):
     self.model = None
     self.graph = dict()
     self.values = list()
     self.dataset = None

  def add(self, vertex_one, vertex_two="None", distance=1):
    if vertex_one not in self.graph.keys():
        self.graph[vertex_one] = []

    self.graph[vertex_one].append(vertex_two)

  def values_to_list(self):
    vals = list()
    for v,k in self.graph.items():
      vals.append(v)
      for j in k:
        vals.append(j)
    self.values = sorted(list(set(vals)))


  def prepare_data(self):
    self.values_to_list()
    size = len(self.values)
    str_to_array = dict()
    for i in range(size):
      arr = np.zeros((size))
      arr[i] = 1
      str_to_array[self.values[i]] = arr

    x_values = list()
    y_values = list()

    for k, v in self.graph.items():
      for i in v:
        x_values.append(str_to_array[k])
        y_values.append(str_to_array[i])

    df = pd.DataFrame()
    df["x_values"] = x_values
    df["y_values"] = y_values

    self.dataset = df

  def fit(self):
    self.prepare_data()
    size = len(self.values)
    self.model = NeuralNetwork(size, size)


    for j in range(5000):
      cost = np.zeros((size, ))
      for i in range(self.dataset.shape[0]):
        pred = self.model.feed_forward(self.dataset.iloc[i, 0])
        y_real = self.dataset.iloc[i, 1]
        loss = pred - y_real
        cost += y_real * np.log(pred)
        self.model.gradient_descent(loss)
      self.model.back_propigate(4e-3)
      cost = -np.sum(cost/size)
      if (cost < 1e-4):
        break

  def predict(self, input):
    index = np.argmax(self.model.feed_forward(input))
    arr = np.zeros((len(self.values), ))
    arr[index] = 1
    return arr
```

```python
g = ProbabilisticGraph()
g.add("a", "b")
g.add("b", "c")
g.add("c", "d")
g.fit()
```
